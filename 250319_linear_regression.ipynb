## 2. 파이썬 코딩으로 확인하는 선형 회귀

### 1.환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. 데이터 준비

```python
# 공부 시간 X와 성적 y 넘파이 배열생성
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

### 3. 데이터 분포 확인

```python
# 데이터를 그래프로
plt.scatter(x, y)
plt.show()
```

### 4. 실행을 위한 변수 설정

```python
# 기울기 a와 절편 b의 값을 초기화
a = 0
b = 0

# 학습률 설정
lr = 0.03

# 반복 수 설정
epochs = 2001
```

### 5. 경사 하강법

```python
# x 값의 개수 세기
n=len(x)

# 경사 하강법
for i in range(epochs):                  # 에포크 수 만큼 반복

    y_pred = a * x + b                   # 예측 값 구하는 식
    error = y - y_pred                   # 실제 값과 비교한 오차를 error로

    a_diff = (2/n) * sum(-x * (error))   # 오차 함수를 a로 편미분한
    b_diff = (2/n) * sum(-(error))       # 오차 함수를 b로 편미분한

    a = a - lr * a_diff     # 학습률을 곱해 원래의 a 값을 업데이트
    b = b - lr * b_diff     # 학습률을 곱해 원래의 b 값을 업데이트

    if i % 100 == 0:        # 100번 반복될 때마다 현재의 a,b 값을 출력
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))
```

### 6. 그래프 확인

```python
# 앞서 구한 최종 a 값을 기울기, b 값을 y 절편에 대입하여 그래프
y_pred = a * x + b

# 그래프 출력
plt.scatter(x, y)
plt.plot(x, y_pred,'r')
plt.show()
```

## 4. 파이썬 코딩으로 확인하는 다중 선형 회귀

### 1.환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. 데이터 준비

```python
# 공부 시간 x1과 과외 시간 x2, 그 성적 y의 넘파이 배열 생성
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

### 3. 데이터 분포 확인

```python
# 데이터의 분포를 그래프
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

### 4. 실행을 위한 변수 설정

```python
# 기울기 a와 절편 b의 값을 초기화
a1 = 0
a2 = 0
b = 0

# 학습률
lr = 0.01

# 몇 번 반복될지 설정
epochs = 2001
```

### 5. 경사 하강법

```python
# x 값의 개수를 센다
n=len(x1)

# 경사 하강법
for i in range(epochs):                  # 에포크 수 만큼 반복

    y_pred = a1 * x1 + a2 * x2 + b       # 예측 값을 구하는 식
    error = y - y_pred                   # 실제 값과 비교한 오차를 error로

    a1_diff = (2/n) * sum(-x1 * (error)) # 오차 함수를 a1로 편미분
    a2_diff = (2/n) * sum(-x2 * (error)) # 오차 함수를 a2로 편미분
    b_diff = (2/n) * sum(-(error))       # 오차 함수를 b로 편미분

    a1 = a1 - lr * a1_diff  # 학습률을 곱해 a1 값을 업데이트
    a2 = a2 - lr * a2_diff  # 학습률을 곱해 a2 값을 업데이트
    b = b - lr * b_diff     # 학습률을 곱해 b 값을 업데이트

    if i % 100 == 0:        # 100번 반복될 때마다 현재의 a1, a2, b 값을 출력
        print("epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f" % (i, a1, a2, b))
```

```python
# 실제 점수와 예측된 점수를 출력
print("실제 점수:", y)
print("예측 점수:", y_pred)
```

## 5-1. 텐서플로에서 실행하는 선형 회귀 모델

### 1. 환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수들을 불러온다
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. 데이터 준비

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

### 3. 모델 실행

```python
model = Sequential()

# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델을 설정
model.add(Dense(1, input_dim=1, activation='linear'))

# 오차 수정을 위해 경사 하강법 오차의 정도를 판단하기 위해 평균 제곱 오차
model.compile(optimizer='sgd', loss='mse')

# 오차를 최소화하는 과정을 2000번 반복
model.fit(x, y, epochs=2000)
```

### 4. 그래프로 확인

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x),'r')    # 예측 결과를 그래프
plt.show()
```

### 5. 모델 테스트

```python
import tensorflow as tf

# 임의의 시간을 집어넣어 점수를 예측하는 모델을 테스트

hour = 7
prediction = model.predict([hour]) #텐서플로 2.9.0 까지만 가능

input_data = tf.constant([[hour]]) # 텐서플로 버전업으로 인한 수정
prediction = model.predict(input_data)[0][0]  # 텐서플로 버전업으로 인한 수정

print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다" % (hour, prediction))
```

## 5-2 텐서플로에서 실행하는 다중 선형 회귀 모델

### 1.환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수들을 불러온다
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### 2. 데이터 준비

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])
```

### 3. 모델 실행

```python
model = Sequential()

# 입력 변수가 2개 input_dim에 2를 입력
model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd' ,loss='mse')

model.fit(x, y, epochs=2000)
```

### 4. 모델 테스트

```python
import tensorflow as tf

# 임의의 학습 시간과 과외 시간을 넣어 점수를 예측하는 모델을 테스트

hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])  #텐서플로 2.9.0 까지만 가능

input_data = tf.constant([[hour, private_class]])  # 텐서플로 버전업으로 인한 수정
prediction = model.predict(input_data)[0][0]  # 텐서플로 버전업으로 인한 수정

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다" % (hour, private_class, prediction))
```
